# ASP.NET Dependecies, Files and Folders
## [Get Start with ASP.NET](https://docs.microsoft.com/en-gb/aspnet/core/tutorials/razor-pages/razor-pages-start?view=aspnetcore-6.0)

1. * # Dependecies
## This folder contains the _framework_ and the _packages_ that are required for the application to run.

2. * # Pages
## Contains all the pages required for the project. Each page is made of two files: 
*  _*.cshtml_  which contains Razor syntax, which is a combination of HTML and C# , 
* _*cshtml.cs_ which holds the logic and behavior for each page.

3. # Properties
## This folder contains a _launchSettings.json_ file, which contains the settings and definitions that are going to be used when running the .NET Core application

* Remeber, This settings will be used only for the local development machine!
* _launchSetting.json_ is only used by Visual Studio


4. # wwwroot
## Contains all the aesthetic files that the application can use. Any HTML, CSS, or Js files will be stored here. We cannot place any Razor or C# files here!
## By default Razor pages use the _CSS_ library called **_bootstrap_** which can be found in this folder under _lib_ directory



5. # appsettings.json
## All application setting will be containing in this file. 
## From a databse connection's string to an access token, ASP.NET will load the value from this file and from the configuration file related to your current environment. 
## Production which describe the environment proved for the customer
## Enviroment settings in which developers work on.


## In the project we find already two environement created for us:


## The _"LogLevel"_ property under _"Logging"_ specifies the minumum _level_ to log for selected categories. By _"Default"_ our application will log anything under the _Information_ categories, which is given when a method end normally.
## Any changes made here will require to restart the application in order to take effect.

5. # Program 
## Is a C#  _*.cs_ file which contains the entry point for the program, **in this file**, the _Main_ method is the first method which gets invoked 

6. # Startup.cs
## Is a C# files which contains the _Startup_ class, where the application's features and services are configured-

```
namespace TestApp
{
    public class Startup
    {
        public Startup(IConfiguration configuration)
        {
            Configuration = configuration;
        }

        public IConfiguration Configuration { get; }

        // This method gets called by the runtime. Use this method to add services to the container.
        public void ConfigureServices(IServiceCollection services)
        {
            services.AddRazorPages();
        }

        // This method gets called by the runtime. Use this method to configure the HTTP request pipeline.
        public void Configure(IApplicationBuilder app, IWebHostEnvironment env)
        {
            if (env.IsDevelopment())
            {
                app.UseDeveloperExceptionPage();
            }
            else
            {
                app.UseExceptionHandler("/Error");
                // The default HSTS value is 30 days. You may want to change this for production scenarios, see https://aka.ms/aspnetcore-hsts.
                app.UseHsts();
            }

            app.UseHttpsRedirection();
            app.UseStaticFiles();

            app.UseRouting();

            app.UseAuthorization();

            app.UseEndpoints(endpoints =>
            {
                endpoints.MapRazorPages();
            });
        }
    }
}
```
